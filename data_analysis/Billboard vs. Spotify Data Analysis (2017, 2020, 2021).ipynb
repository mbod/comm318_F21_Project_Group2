## Billboard vs. Spotify Data Analysis (2017, 2020, 2021)

Comparing Spotify's chart performance data with Billboard's chart performance data (paired with Spotify's track-level audio metrics) in order contextualize and analyze the Billboard Year-End Top 100 Songs from 2017, 2020, and 2021.

Source: Spotify Web API, Billboard Web API, [Spotify Top Tracks of 2020 USA](https://open.spotify.com/user/spotify/playlist/37i9dQZF1DXaqCgtv7ZR3L?si=eAq2hBqrTR-s5M99y-vQeQ), [Spotify Top Tracks of 2017: USA](https://open.spotify.com/user/spotify/playlist/37i9dQZF1DX7Axsg3uaDZb?si=IAEexkHXRTGnpkaGpZ0Fxw), [Spotify Top Tracks of 2021 USA](https://open.spotify.com/user/spotify/playlist/37i9dQZF1DXbJMiQ53rTyJ?si=GBsy_cPARumBOd1C_2es2w), Billboard Hot 100 in [2017](https://open.spotify.com/user/whe1998/playlist/255aUSCuVTcdD5JTogG69d?si=JCHWINVIRPiW3K_aGVzgcg), [2020](https://open.spotify.com/user/zoscar_/playlist/1WBljFutuk7uLQtfqfmjWV?si=DnFDuF-0SRuIfMVzPC6hOg), and [2021](https://open.spotify.com/user/11148724827/playlist/5Nt7KFSEfXIlsDIB8SCpNU?si=FuMmkxzlTKqHo8Vf9-BXtA).

```python
# Importing modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
# Loading in data
billboard2017_clean = pd.read_csv('../data/Clean/billboard2017cleaned.csv')
billboard2020_clean = pd.read_csv('../data/Clean/billboard2020cleaned.csv')
billboard2021_clean = pd.read_csv('../data/Clean/billboard2021cleaned.csv')
spotify2017_clean = pd.read_csv('../data/Clean/spotifytop2017cleaned.csv')
spotify2020_clean = pd.read_csv('../data/Clean/spotifytop2020cleaned.csv')
spotify2021_clean = pd.read_csv('../data/Clean/spotifytop2021cleaned.csv')
```

#### Question 1: How do danceability and energy compare between Spotify and Billboard charts?

```python
## Comparing mean danceability/energy between 2017 Spotify/Billboard year-end charts

# Computing means
spotify2017_mean_dance = spotify2017_clean['danceability'].mean()
spotify2017_mean_energy = spotify2017_clean['energy'].mean()
billboard2017_mean_dance = billboard2017_clean['danceability'].mean()
billboard2017_mean_energy = billboard2017_clean['energy'].mean()

# Creating new dataframe and plotting
dance_2017 = [spotify2017_mean_dance, billboard2017_mean_dance]
energy_2017 = [spotify2017_mean_energy, billboard2017_mean_energy]
index = ['Spotify', 'Billboard']
de2017_mean_df = pd.DataFrame({'danceability' : dance_2017, 'energy' : energy_2017}, index = index)

de2017_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Danceability/Energy: Spotify vs. Billboard in 2017')
plt.ylabel('Mean Ratings')
plt.show()
```

Look at the similarity in heights between the blue bars and the orange bars, there's not too much difference for the 2017 lists - this is in line with what I would've predicted, as I don't think the streaming-exclusive charts started deviating from the Billboard charts until more recent years. Let's try 2020:

```python
## Comparing mean danceability/energy between 2020 Spotify/Billboard year-end charts

# Computing means
spotify2020_mean_dance = spotify2020_clean['danceability'].mean()
spotify2020_mean_energy = spotify2020_clean['energy'].mean()
billboard2020_mean_dance = billboard2020_clean['danceability'].mean()
billboard2020_mean_energy = billboard2020_clean['energy'].mean()

# Creating new dataframe and plotting
dance_2020 = [spotify2020_mean_dance, billboard2020_mean_dance]
energy_2020 = [spotify2020_mean_energy, billboard2020_mean_energy]
index = ['Spotify', 'Billboard']
de2020_mean_df = pd.DataFrame({'danceability' : dance_2020, 'energy' : energy_2020}, index = index)

de2020_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Danceability/Energy: Spotify vs. Billboard in 2020')
plt.ylabel('Mean Ratings')
plt.show()
```

As it turns out, the 2020 charts were also pretty similar in terms of danceability and energy. I would've predicted that Spotify charting songs would have higher danceability and energy, but perhaps the overlap between the two charts is still too strong for there to be a noticeable difference.

```python
# Computing means in 2021
spotify2021_mean_dance = spotify2021_clean['danceability'].mean()
spotify2021_mean_energy = spotify2021_clean['energy'].mean()
billboard2021_mean_dance = billboard2021_clean['danceability'].mean()
billboard2021_mean_energy = billboard2021_clean['energy'].mean()

# Creating new dataframe and plotting object
dance_2021 = [spotify2021_mean_dance, billboard2021_mean_dance]
energy_2021 = [spotify2021_mean_energy, billboard2021_mean_energy]
index = ['Spotify', 'Billboard']
de2021_mean_df = pd.DataFrame({'danceability' : dance_2021, 'energy' : energy_2021}, index = index)

de2021_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Danceability/Energy: Spotify vs. Billboard in 2021')
plt.ylabel('Mean Ratings')
plt.show()
```

Again, there are barely any noticeable differences in danceability or energy between Spotify and Billboard in 2021.

#### Question 2: How do song lengths compare between Spotify and Billboard charts?

```python
## Comparing mean song lengths between 2017 Spotify/Billboard year-end charts

# Computing means
spotify2017_mean_duration = spotify2017_clean['duration_sec'].mean()
billboard2017_mean_duration = billboard2017_clean['duration_sec'].mean()

# Creating new dataframe and plotting
duration_2017 = [spotify2017_mean_duration, billboard2017_mean_duration]
index = ['Spotify', 'Billboard']
duration2017_mean_df = pd.DataFrame({'duration_sec' : duration_2017}, index = index)

duration2017_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Song Length: Spotify vs. Billboard in 2017')
plt.ylabel('Song Length (sec)')
plt.show()
```

```python
## Comparing mean song lengths between 2020 Spotify/Billboard year-end charts

# Computing means
spotify2020_mean_duration = spotify2020_clean['duration_sec'].mean()
billboard2020_mean_duration = billboard2020_clean['duration_sec'].mean()

# Creating new dataframe and plotting
duration_2020 = [spotify2020_mean_duration, billboard2020_mean_duration]
index = ['Spotify', 'Billboard']
duration2020_mean_df = pd.DataFrame({'duration_sec' : duration_2020}, index = index)

duration2020_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Song Length: Spotify vs. Billboard in 2020')
plt.ylabel('Song Length (sec)')
plt.show()
```

```python
## Comparing mean song lengths between 2021 Spotify/Billboard year-end charts

# Computing means
spotify2021_mean_duration = spotify2021_clean['duration_sec'].mean()
billboard2021_mean_duration = billboard2021_clean['duration_sec'].mean()

# Creating new dataframe and plotting
duration_2021 = [spotify2021_mean_duration, billboard2021_mean_duration]
index = ['Spotify', 'Billboard']
duration2021_mean_df = pd.DataFrame({'duration_sec' : duration_2021}, index = index)

duration2021_mean_df.plot.bar(rot = 0)
plt.legend(loc = 'lower right')
plt.title('Mean Song Length: Spotify vs. Billboard in 2021')
plt.ylabel('Song Length (sec)')
plt.show()
```

```python
# Computing means
spotify2017_mean_duration = spotify2017_clean['duration_sec'].mean()
billboard2017_mean_duration = billboard2017_clean['duration_sec'].mean()

spotify2020_mean_duration = spotify2020_clean['duration_sec'].mean()
billboard2020_mean_duration = billboard2020_clean['duration_sec'].mean()

spotify2021_mean_duration = spotify2021_clean['duration_sec'].mean()
billboard2021_mean_duration = billboard2021_clean['duration_sec'].mean()

# Create data
data = ["Spotify", spotify2017_clean['duration_sec'].mean(), spotify2020_clean['duration_sec'].mean(), spotify2021_clean['duration_sec'].mean()], ["Billboard", billboard2017_clean['duration_sec'].mean(), billboard2020_clean['duration_sec'].mean(), billboard2021_clean['duration_sec'].mean()]

# Display table
print(tabulate(data, headers = ["Chart Mean Duration (sec)", "2017", "2020",  "2021"]))
```

In all our plots and tables, there are barely any noticeable difference in mean song length. Even though it does seem that songs in general across charts are getting shorter over the years, song duration between Spotify and Billboard within each year seems to be the same.
